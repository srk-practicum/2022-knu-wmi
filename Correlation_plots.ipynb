```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

csi_matrix = pd.DataFrame(pd.read_csv("csv_files/falldown/Falldown_1.csv"))
plt.matshow(csi_matrix.corr())
plt.show()

corr = csi_matrix.corr()
ax = sns.heatmap(
    corr,
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True)
```
