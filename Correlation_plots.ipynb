```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
main_dir = os.listdir("csv_training")
for ind, i in enumerate(main_dir):
    files_list = os.listdir(f"csv_training/{i}")
    for j in files_list:
        csi_matrix = pd.DataFrame(pd.read_csv(f"csv_training/{i}/{j}"))
        plt.matshow(csi_matrix.corr())
        plt.title(f"{j}")
        plt.show()
```
