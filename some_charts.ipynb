```python
# csi_matrix = pd.DataFrame(pd.read_csv("csv_test/falldown/Falldown_1.csv"))
packages = []
dir_csv = os.listdir("csv_test")
for ind, i in enumerate(dir_csv):
    files_list = os.listdir(f"csv_test/{i}")
    counter = 1
    for j in files_list:
        csi_matrix = pd.DataFrame(pd.read_csv("csv_test/falldown/Falldown_1.csv"))
        for k in range (1,len(csi_matrix)):

            csi_matrix.iloc[k].plot.line(legend=False)
    csi_matrix.iloc[k].plot.line(legend=False,title = j)

# packages.plot.line(legend=False)
```
