```python
import matplotlib.pyplot as plt
import pandas as pd
from CSIKit.tools.batch_graph import BatchGraph
import os
```

## DotPlot

```python
colors = ['yellow', 'blue', 'orange', 'black', 'purple', 'brown', 'green']
csv_folders = os.listdir("csv_folders")
plt.figure(figsize=(8, 3))
for ind, folder in enumerate(csv_folders):
    files = os.listdir(f"csv_folders/{folder}")
    ctr = 1
    for file in files:
        length = len(pd.DataFrame(pd.read_csv(f"csv_folders/{folder}/{file}")))
        ctr += 1
        plt.scatter(length, ctr, color=colors[ind])
    plt.scatter(length, ctr, color=colors[ind],label=csv_folders[ind])
plt.legend()
plt.gca().set(xlim=(24, 350), ylim=(0, 15), xlabel='Packets', ylabel='Count')
plt.grid()

```
