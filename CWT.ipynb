# CWT Heatmap :

```python
N = 10
L = 30
frequencies = [2 ** (n + (l / L)) for n in range(N) for l in range(L)]
#dt = 0.003 test variable
#frequencies = scale2frequency('morl', np.arange(1,50,2.5)) / dt
#frequencies = np.flip(frequencies)
```

```python
#Command dataset
#df = pd.read_csv(mypath + "Falldown_1.csv")
#Internet dataset
df = pd.read_csv(mypath + "sleeping13.csv")
coef,freqs = cwt(df['1'],frequencies,wavelet='morl')
sns.heatmap(coef,xticklabels=False, yticklabels=False,cmap="viridis")
plt.show()
```
