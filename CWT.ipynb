```python
from pywt import cwt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(mypath + "Falldown_1.csv")
coef,freqs = cwt(df['15'],np.arange(1,400),wavelet='morl')
sns.heatmap(coef,xticklabels=False, yticklabels=False,cmap="viridis")
plt.show()
```
