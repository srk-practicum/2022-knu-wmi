# Imports :

```python
import numpy as np
import pandas as pd
from pywt import cwt
from os import listdir
from os.path import isfile, join
import seaborn as sns
import matplotlib.pyplot as plt
```

# All Files Names : 

```python
mypath = r"pca/pca_csv_phase/"
filenames = [f.split(".")[0] for f in listdir(mypath) if isfile(join(mypath, f))]
```

# CWT Heatmap :

```python
N = 10
L = 30
scales = [2 ** (n + (l / L)) for n in range(N) for l in range(L)]
```

```python
df = pd.read_csv(mypath + "Falldown_1.csv")
coef,freqs = cwt(df['15'],scales,wavelet='morl')
sns.heatmap(coef,xticklabels=False, yticklabels=False,cmap="viridis")
plt.show()
```
