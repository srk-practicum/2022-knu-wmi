```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pywt
import pywt.data

files_csv_train_PCA = []
for i in os.listdir('files_csv_train_PCA'):
    for j in os.listdir(f'files_csv_train_PCA/{i}'):
        df = pd.read_csv(f'files_csv_train_PCA/{i}/{j}')
        files_csv_train_PCA.append(df)

def wavelets1():
    for i in files_csv_train_PCA:
        coef,freqs = pywt.cwt(i[0],np.arange(1,300),wavelet='morl')
        sns.heatmap(coef,cmap='jet_r', shading='gouraud')
        plt.show()
def wavelets2():
    for i in files_csv_train_PCA:
        for j in i.keys():
            coef,freqs = pywt.cwt(i[f"{j}"],np.arange(1,300),wavelet='morl')
            sns.heatmap(coef,cmap='jet_r', shading='gouraud')
            plt.show()
#wavelets2()
```
