```python
import matplotlib.pyplot as plt
import os
import pandas as pd

plt.style.use('seaborn-whitegrid')
markers = ['^', 'v', 'o', '<', '>', 's', 'd']
colors = ['brown', 'green', 'blue', 'red', 'purple', 'orange', 'black']
numlines = [[], [], [], [], [], [], []]
numlines1 = []
dir_csv = os.listdir("files_csv")
k = 0
for i in dir_csv:
    files_list = os.listdir(f"files_csv/{i}")
    for j in files_list:
        numline = len(pd.DataFrame(pd.read_csv(f"files_csv/{i}/{j}")))
        numlines[k].append(numline)
        numlines1.append(numline)
    k += 1

k = 0
index = -1
for i in range(len(numlines)):
    for j in range(len(numlines[i])):
        index += 1
        a = numlines[i][j]
        count = 1
        for p in range(len(numlines1)):
            if (a == numlines1[p] and p != index):
                count += 1
        plt.plot(a, count, markers[k], color = colors[k])
    plt.plot(a, count, markers[k], color = colors[k], label=dir_csv[i])
    k += 1

plt.legend()
plt.show()
```
