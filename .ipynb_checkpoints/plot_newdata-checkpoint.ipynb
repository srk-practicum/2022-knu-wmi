```python
from CSIKit.reader import get_reader
from CSIKit.util import csitools
from CSIKit.tools.batch_graph import BatchGraph

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import os
from collections import Counter
```

## DotPlot

```python
plt.figure(figsize=(8, 5))

directory = os.listdir('data_csi')
frames = Counter()
renders = 0
count = 0
marker = ['s', '<', '^', 'x', '+', 'v', '^', '>', '1', '2', '3', '4']
marker_type = ''

for file in directory:
    renders += 1
    my_reader = get_reader(f"data_csi/{file}")
    csi_data = my_reader.read_file(f"data_csi/{file}",scaled=True)
    csi_amplitude, no_frames, no_subcarriers = csitools.get_CSI(csi_data, 'amplitude')

    frames[no_frames] = frames[no_frames]+ 1 if no_frames in frames else  1

    if renders == 1:
        marker_type = marker[count]
        plt.plot(frames.keys(), frames.values(), marker_type, label=f"{file.strip('(1).pcap')}")

    elif renders == 30:
        renders = 0
        count += 1
        plt.plot(frames.keys(), frames.values(), marker_type)

    else:
        plt.plot(frames.keys(), frames.values(), marker_type)
        
    plt.legend()


plt.xlabel('Number of Frames')
plt.ylabel('Amount')
```
