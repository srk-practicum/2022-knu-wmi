```python
import sys
import pandas as pd
import pylab as pl
from sklearn import preprocessing
from sklearn.decomposition import PCA
import csv
import pywt
import  numpy as np
import matplotlib.pyplot as plt
```

```python
window_size = 50
N = 7
L = 15
scales = [2 ** (n + (l / L)) for n in range(N) for l in range(L)]
"""Load data."""
csvfile = "csv_files/walk/walk_1591628660.csv"
data = pd.read_csv(csvfile, index_col=(0, 1))
# first column provides labels
xdata = data.values
pca = PCA(n_components=150)
components = pca.fit(xdata).transform(xdata)
# Percentage of variance explained for each components
print('explained variance ratio (first two components):',
        pca.explained_variance_ratio_)
x = 5
writer = csv.writer(open("csvfile_pca.csv", 'w'))
writer.writerows(components)
coef, freqs = pywt.cwt(components[x], scales, 'morl')
plt.matshow(coef) # doctest: +SKIP
plt.show() # doctest: +SKIP
```
