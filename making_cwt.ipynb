```python
import os
import sys
import pandas as pd
from sklearn import preprocessing
from sklearn.decomposition import PCA
import csv
import pywt
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
window_size = 50
N = 7
L = 15
scales = [2 ** (n + (l / L)) for n in range(N) for l in range(L)]
# """Load data."""
# csvfile = "csv_training_files_converted/brushteeth/brushteeth_1591629519.csv"
# data = pd.DataFrame(pd.read_csv(csvfile, index_col=(0, 1)))
# print(data)
# # first column provides labels
dirs = os.listdir("csv_training_not_converted")
for i in dirs :
    files_list = os.listdir(f"csv_training_not_converted/{i}")
    files = []

    for ind,j in enumerate (files_list) :
        data = pd.read_csv(f"csv_training_not_converted/{i}/{j}", index_col=(0, 1))
        # first column provides labels
        xdata = data.values
        pca = PCA(n_components=7)

        components = pca.fit(xdata).transform(xdata)
        # print(components)
        # Percentage of variance explained for each components
        # print('explained variance ratio (first two components):',
        #       pca.explained_variance_ratio_)
        components = pd.DataFrame(components)
        files.append(components)
        components[0].plot.line(legend = False)
    df = pd.concat(files, ignore_index=True)
    # print(df)
    x = 5
    print(f"csv_training_not_converted/{i}")
    # writer.writerows(components)
    coef, freqs = pywt.cwt(xdata[x], scales, 'morl')
    plt.matshow(coef)
    plt.title(f"csv_training_not_converted/{i}")# doctest: +SKIP
    plt.colorbar()
    plt.show() # doctest: +SKIP
```
